```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import joblib

# 합성 데이터셋 생성 함수 정의
def generate_road_dataset(num_samples=5000):
    data = []
    # 연/월/일 제외하고 '시간'만 표시 (00:00 ~ 23:00 반복)
    for i in range(num_samples):
        hour = i % 24
        time_str = f"{hour:02d}:00"

        # 상태 결정 (안전, 눈길, 빗길)
        state_roll = random.random()

        if state_roll < 0.6:  # 안전 (60%)
            label = "안전"
            conductivity = random.randint(850, 1000)
            temp = round(random.uniform(5.0, 35.0), 1)
            humidity = random.randint(20, 50)

        elif state_roll < 0.8:  # 눈길 (20%)
            label = "눈길(위험)"
            conductivity = random.randint(150, 850)
            # 대부분 영하, 간혹 0~3도
            temp = round(random.uniform(-15.0, 3.0), 1)
            humidity = random.randint(30, 65)

        else:  # 빗길 (20%)
            label = "빗길(조심)"
            conductivity = random.randint(0, 100)
            temp = round(random.uniform(4.0, 25.0), 1)
            humidity = random.randint(75, 99)

        data.append([time_str, temp, humidity, conductivity, label])

    df = pd.DataFrame(data, columns=['time', 'temperature', 'humidity', 'conductivity', 'label'])
    return df

# 데이터 생성 및 CSV 저장 5000개
print("데이터 생성 중...")
dataset = generate_road_dataset(5000)

# 코랩 환경의 현재 디렉토리에 저장
file_name = 'road_dataset.csv'
dataset.to_csv(file_name, index=False, encoding='utf-8-sig') # 파이썬에서 데이터를 저장하기 위해 CSV파일을 생성하고 엑셀과 같은 파일로 데이터를 확인할 때 한칸씩 밀린다면 encoding을 utf-8에서 utf-8-sig로 해결.
print(f"데이터셋 저장 완료: {file_name}")

# (선택 사항) 생성된 데이터 샘플 확인
print("\n[데이터셋 상위 5개 샘플]")
print(dataset.head())
```
